```python
import requests,bs4
```

```python

covid_data_ng = requests.get("https://en.wikipedia.org/wiki/COVID-19_pandemic_in_Nigeria")
soup = bs4.BeautifulSoup(markup= covid_data_ng.content, features="html.parser" )
covid_table_ng = soup.find('table',{'class':'wikitable sortable'})
covid_table_ng
```

```python
tbody = covid_table_ng.tbody
```

```python
tr_elements = tbody.find_all("tr")
for element in tr_elements:
    print(element.th.text)
    try:
        print(element.find_all("td")[0].text)
        print(element.find_all("td")[1].text)
        print(element.find_all("td")[2].text)
    except:
        pass
    print()
```
